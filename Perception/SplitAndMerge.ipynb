```python
def find_line_model(x1,y1,x2,y2):
    if (x2-x1==0):
      return math.inf,math.inf
    m = (y2-y1) / (x2-x1) 
    c = y1 - m * x1     
    return m, c
```

```python
def find_intercept_point(m, c, x0, y0):
    x = (x0 + m*y0 - m*c)/(1 + m**2)
    y = (m*x0 + (m**2)*y0 - (m**2)*c)/(1 + m**2) + c
    return x, y
```
